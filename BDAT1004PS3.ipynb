```python
## Change the name of the columns to bedrs, bathrs, price_sqr_meter

df = df.rename(columns = {'a' : 'bedrs', 'b' : 'bathrs', 'c' : 'price_sqr_meter'}, inplace = False)
df
```

```python
'''Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!'''
import seaborn as sns
import matplotlib.pyplot as plt
mar_div = pd.read_csv('us-marriages-divorces-1867-2014.csv')
mar_div.head()
```

```python
columns = mar_div[['Marriages','Divorces']]

x_data = mar_div[['Year']]

# create figure and axis
fig, ax = plt.subplots()


for column in columns:
    ax.plot(x_data, mar_div[column])
ax.legend(columns)
ax.set_ylabel('Number of marriages & divorces')
ax.set_xlabel('Year')
plt.show()
```

```python
'''Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!'''

comp = mar_div[(mar_div['Year'] == 1900) | (mar_div['Year'] == 1950)| (mar_div['Year'] == 2000)][['Year','Marriages','Divorces']].plot.bar('Year')

comp.set_xlabel('Year')
comp.set_ylabel('number of marriages and divorces')
plt.show()
```
